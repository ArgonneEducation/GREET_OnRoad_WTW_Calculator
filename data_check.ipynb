```python
import pandas as pd
```

```python
import json

# Load the JSON file
with open('emissions.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data
df = pd.json_normalize(data, sep='_', errors='ignore')

df.head(10)
```

```python
# Remove the first two arrays from the JSON data
data_without_arrays = {key: value for key, value in data.items() if key not in ['feedstocks', 'metrics']}

# Normalize the modified JSON data
df = pd.json_normalize(data_without_arrays['emissions'], sep='_', errors='ignore')

df.head(10)
```

```python
df.set_index('id', inplace=True)
```

```python
df_transposed = df.T
df_transposed.head(32)
```

Note: the head of the dataframe shows all data through BC_Total, the tail of the dataframe, shown below, shows all of the remaining data, with the BC_Total units included at the first row as a reference to show no missing rows.

```python
import csv

def add_text_to_csv(file_path, text):
    # Read the original content of the CSV file
    with open(file_path, 'r', newline='') as file:
        reader = csv.reader(file)
        rows = list(reader)

    # Modify the first row by adding the text before the first comma
    if rows:
        first_row = rows[0]
        if first_row:
            first_row[0] = text + '' + first_row[0]

    # Write the modified content back to the CSV file
    with open(file_path, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)

# Example usage
file_path = 'cleaned_gas_data.csv'
add_text_to_csv(file_path, 'Metric_Stage')
```
